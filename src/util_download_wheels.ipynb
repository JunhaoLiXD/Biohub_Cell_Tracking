# Download zarr wheels for offline submission
Run this with **Internet ON**. It downloads zarr>=3 (+ all dependencies) as wheels, verifies they install offline, then you turn the `wheels/` folder into a Kaggle Dataset named `zarr3-offline`.

**After running:** Save Version -> Output -> New Dataset from the `wheels` folder -> name it `zarr3-offline`. Attach that dataset to the submission notebook.

```python
# 1) Download zarr>=3 and ALL its dependencies as wheels into /kaggle/working/wheels
import os, glob, subprocess, sys, shutil
WHEELS = '/kaggle/working/wheels'
shutil.rmtree(WHEELS, ignore_errors=True)
os.makedirs(WHEELS, exist_ok=True)
r = subprocess.run([sys.executable, '-m', 'pip', 'download', 'zarr>=3', '-d', WHEELS],
                   capture_output=True, text=True)
print(r.stdout[-2000:])
if r.returncode: print('STDERR:\n', r.stderr[-2000:])
files = sorted(os.listdir(WHEELS))
print(f'\n{len(files)} files downloaded:')
for f in files: print('  ', f)
```

```python
# 2) VERIFY the wheels install fully offline (no network) -- this proves the set is self-contained.
#    --no-index forbids PyPI, --find-links uses only the local wheels dir.
r = subprocess.run([sys.executable, '-m', 'pip', 'install',
                    '--no-index', '--find-links', WHEELS, 'zarr'],
                   capture_output=True, text=True)
print(r.stdout[-2000:])
if r.returncode:
    print('OFFLINE INSTALL FAILED:\n', r.stderr[-2000:])
else:
    import importlib, zarr; importlib.reload(zarr)
    print('\nOK -> offline install works. zarr', zarr.__version__)
    print('This is exactly what the submission notebook will do (with internet OFF).')
```

```python
# 3) Next steps (manual):
print('Save Version, then create a Kaggle Dataset from the folder:')
print('   ', WHEELS)
print('Name it:  zarr3-offline')
print('Then attach zarr3-offline as input to submit.ipynb, turn Internet OFF, Run All to verify.')
```
